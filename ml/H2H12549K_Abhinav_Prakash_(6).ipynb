```python
Q1 = df['AMOUNT_IN_USD'].quantile(0.25)
Q3 = df['AMOUNT_IN_USD'].quantile(0.75)
IQR = Q3 - Q1

# Define the threshold for outliers
threshold = 1.5
# Find the outliers
outliers = (df['AMOUNT_IN_USD'] < Q1 - threshold * IQR) | (df['AMOUNT_IN_USD'] > Q3 + threshold * IQR)
print(outliers)
```

```python
# Replacing the outliers with median value
df.loc[outliers, 'AMOUNT_IN_USD'] = df['AMOUNT_IN_USD'].median()
df.AMOUNT_IN_USD
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
#defining a function for performing the hyperparamter tuning on random forest regressor.
def ModelSelection(list_of_models, hyperparameters_dictionary):

  result = []

  i = 0

  for model in list_of_models:

    key = model_keys[i]

    params = hyperparameters_dictionary[key]

    i += 1

    print(model)
    print(params)
    print('---------------------------------')


    classifier =RandomizedSearchCV(model, params, cv=5)

    # fitting the data to classifier
    classifier.fit(X_train,y_train)

    result.append({
        'model used' : model,
        'highest score' : classifier.best_score_,
        'best hyperparameters' : classifier.best_params_
    })

  result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])

  return result_dataframe
```

```python
ModelSelection(models, model_hyperparameters)
```
